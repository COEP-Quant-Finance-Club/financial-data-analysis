```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
stocks_to_work_df = pd.read_csv('sorted_stocks_by_marketcap.csv')
```

```python
stocks_to_work_df.columns = stocks_to_work_df.columns.str.strip()
stocks_to_work = stocks_to_work_df['Symbol'].iloc[401:801]
```

```python
stocks_to_work = stocks_to_work.str.strip()
stocks_to_work.iloc[0]
```

```python
stocks_to_work = stocks_to_work + '.NS'
stocks_to_work
```

```python
data = yf.download(
    list(stocks_to_work),
    start="2021-01-01",
    group_by="ticker",
    auto_adjust=True
)
```

```python
data
```

```python
import os

output_dir = "ohlcv_data"
os.makedirs(output_dir, exist_ok=True)

ohlcv_cols = ["Open", "High", "Low", "Close", "Volume"]

for ticker in data.columns.levels[0]:
    df = data[ticker][ohlcv_cols].dropna()
    df.to_csv(f"{output_dir}/{ticker}.csv")
```
