```python
import yfinance as yf
import pandas as pd
import os
import time






df_list = pd.read_csv('sorted_stocks_by_marketcap_800_to_1000_new.csv')


df_list.columns = df_list.columns.str.strip()


symbols = (
    df_list['Symbol']
    .dropna()                            
    .astype(str)                         
    .str.strip()                         
    .str.replace(r'^\$', '', regex=True) 
    .tolist()
)

# Final filter: Keep only actual text strings
symbols = [s for s in symbols if len(s) > 0 and s != 'nan']

print(f"Verified {len(symbols)} clean symbols. Starting download...")



folder_name = 'Adjusted_Close_for_Drawdown/Ulcer'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    print(f"Created folder: {folder_name}")



print(f"Starting download for {len(symbols)} stocks...")

for i, ticker in enumerate(symbols):
    ticker_ns = f"{ticker}.NS"
    try:
        
        df = yf.download(ticker_ns, start="2021-01-01", end="2026-01-15", auto_adjust=True, progress=False)
        if not df.empty:
            if isinstance(df.columns, pd.MultiIndex):
                df.columns = df.columns.get_level_values(0)
            df = df[['Close']]
            df = df.reset_index()
            file_path = os.path.join(folder_name, f"{ticker}.csv")
            df.to_csv(file_path, index=False)
            
            print(f"[{i+1}/200] Saved Adjusted Close: {file_path}")
        else:
            print(f"⚠️ [{i+1}/200] No data found for {ticker}")
            
    except Exception as e:
        print(f"❌ Error downloading {ticker}: {e}")

    
    time.sleep(0.5)

print("\n✅ All downloads complete. Check the 'Adjusted_Close_for_Drawdown/Ulcer' folder.")
```
