# Sorting top 1000 Stocks

Sorting the 2500 stocks in marketcap_2500list.csv into 

```python
import yfinance as yf
import pandas as pd
from time import sleep

df = pd.read_csv("marketcap_2500list.csv")

# Strip whitespace from column names
df.columns = df.columns.str.strip()

market_caps = []
total = len(df)
errors = 0

# Extract symbols and fetch market cap data
for idx, row in df.iterrows():
    if (idx + 1) % 50 == 0:
        print(f"Progress: {idx + 1}/{total} stocks processed... ({errors} errors so far)")

    symbol = row['SYMBOL'].strip()  # Strip whitespace from symbol
    company = row['NAME OF COMPANY']

    try:
        ticker_symbol = symbol + ".NS"
        ticker = yf.Ticker(ticker_symbol)
        market_cap = ticker.info.get('marketCap', None)

        market_caps.append({
            'Symbol': symbol,
            'Company': company,
            'Market Cap': market_cap
        })
    except Exception as e:
        errors += 1
        print(f"Error fetching {ticker_symbol}: {e}")
        market_caps.append({
            'Symbol': symbol,
            'Company': company,
            'Market Cap': None
        })

    # Reduced delay to avoid rate limiting (was 0.2, now 0.05)
    sleep(0.2)

print(f"Completed: {total}/{total} stocks processed with {errors} errors.")

# Create dataframe with results
market_cap_df = pd.DataFrame(market_caps)

# Sort by market cap in descending order (highest to lowest)
market_cap_df_sorted = market_cap_df.sort_values(by='Market Cap', ascending=False)

# Save sorted results to CSV
market_cap_df_sorted.to_csv('sorted_stocks_by_marketcap.csv', index=False)

print(f"\nTotal stocks processed: {len(market_cap_df_sorted)}")
print(f"Stocks with market cap data: {market_cap_df_sorted['Market Cap'].notna().sum()}")
print("\nTop 10 stocks by market cap:")
print(market_cap_df_sorted.head(10))

market_cap_df_sorted
```
